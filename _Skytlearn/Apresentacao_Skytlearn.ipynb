# Tipos de aprendizado de maquina

O aprendizado de maquina pode ser supervisionado ou não supervisionado

- **Aprendizado supervisionado VS Não supervisionado**

O aprendizado supervisionado e não supervisionado possui uma diferença significativa. O aprendizado supervisionado utiliza conjuntos de dados rotulados, enquanto o não supervisionado emprega conjuntos de dados não rotulados. O termo “rotulado” significa que os dados já estão marcados com a resposta correta.


https://www.alteryx.com/pt-br/glossary/supervised-vs-unsupervised-learning#:~:text=O%20aprendizado%20supervisionado%20e%20n%C3%A3o,marcados%20com%20a%20resposta%20correta.


## Aprendizado supervisionado

No aprendizado supervisionado, a base de treinamento é fornecido ao algoritmo de Machine Learning contendo tanto as entradas quantos os rotulos (Labels) das saídas. Podemos dividir o aprendizados supervisionado em dois tipos:

- Regressão

Na regressão, a saída gerada pelo modelo é um valor contínuo. Com modelos de regressão, é possível fazer previsões de causa e efeito com base em múltiplos pontos de dados. Em uma empresa, por exemplo, isso pode incluir a projeção do crescimento da receita publicitária. Entre alguns algoritmos de regressão comuns estão a regressão de ridge, lasso, regressão de redes neurais e regressão logística.

- Classificação

Na classificação, a saída gerada pelo modelo é um valor pertencente a um conjunto discreto, podendo ser **categoria** ou **labels**. Um exemplo prático é um algoritmo que ajuda a eliminar spams na caixa de entrada prioritária de e-mails ou um algoritmo que ajuda usuários a bloquear ou restringir o acesso de outras pessoas nas redes sociais. Alguns algoritmos de classificação comuns incluem regressão logística, k-vizinhos mais próximos, floresta aleatória, Naive Bayes, gradiente descendente estocástico e árvores de decisão.

### Preprocessamento com o Skit-learn



#### Classe e metodos LabelEncoder (transformando dados categoricos ordinais ou nominais em numeros entre zero e um)




```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
```

- `preprocessing.LabelEncoder()` Armazena numeros entre 0 e n_classes-1 , sendo utilizado para armazerar dados categoricos de forma numerica no seu atributo <span style="color:#0074D9"> classes_ </span>  
<br>
- `le.fit([l])` armazena os valores unicos da lista l,criando assim as n_classes de dados categoricos (numericos ou não)
- `le.transform([l])` transforma uma lista l contendo  alguma das n_classes de dados categoricos (numericos ou não) num conjunto de numeros entre 0 e n_classes-1 e armazena os numeros no Label Encoder
- `le.fit_transform([l])` realiza a etapa de fit e transformação num unico metodo
- `le.classes_` retorna um array com as classes salvas no LabelEncoder le


https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html#sklearn.preprocessing.LabelEncoder

Para dados categoricos não binario como por exemplo, manhã, tarde e noite , pode ser necessrio binarizar os mesmos criando por exemplo uma coluna manhã que pode assumir 0 ou 1 , para esses casos tem-se odis metodos do pandas abaixo que possibilitam realizar essa binarização

```python
import pandas as pd

```

- `pd.get_dummies(df,columns=['Collumns']))` retorna um dataframe com colunas novas para cada um dos valores possiveis das colunas Colluumns, serve para binarizar atributos categoricos

Para mais detalhes sobre pandas vide [_Numpy_e_Pandas](../_Numpy_e_Pandas/Numpy_e_Pandas.ipynb)

````python
# Importando do Dataframe para demonstrar os metodos do preprocessing do sckitlearn
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv')
````

````python
#apenas verificando que a coluna sexo não tem valores nulos para podermos iniciar o proprocessamento da coluna sexo
df['Sex'].isna().value_counts()
````

````python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()  

# Salva todas as classes existentes no df['Sex']
le.fit(df['Sex'])
````

````python
le.classes_
````

````python
list(le.classes_)
````

````python
#Cria uma variavel categorica Sex_le para substituir as categorias female e male e assim poder realizar o aprendizado de maquina com variaveis categoricas numericas

df['Sex_le'] = le.transform(df['Sex'])
````

Ao inves do skit-learn para criar uma variavel categorica numerica para o sexo, seria possivel realizar a mesma transformação com mais linhas de codigo sem o uso do skit-learn

```python
df.loc[df['Sex']=='male', 'Sex_le'] = 1
df.loc[df['Sex']=='female', 'Sex_le'] = 0

```

````python

#remoção de campos desnecessarios

df.drop(columns=['PassengerId','Name','Sex', 'Ticket', 'Cabin'], inplace=True)
df.head()
````

````python
#apenas verificando que a coluna embarque não tem valores nulos para podermos iniciar o proprocessamento da coluna embarque
df['Embarked'].isna().value_counts()
````

````python

#Binarizando o tipo de embarque, os valores que eram NAN serão definidos como Zero ou False apos a binarização

df=pd.get_dummies(df,columns=['Embarked'])
````

````python
df.info()
````

````python
#transformando embarked_Q, C e S em valores numericos, ao realisar o pd.get_dummies no colab não precisou dessa etapa


df['Embarked_Q']=df['Embarked_Q'].astype('int')
df['Embarked_S']=df['Embarked_S'].astype('int')
df['Embarked_C']=df['Embarked_C'].astype('int')
````

Tambem poderia ter sido transformado os valores boleano em numericos com a função loc

```python
df.loc[df['Embarked_Q']==True,'Embarked_Q']= 1
df.loc[df['Embarked_Q']==False,'Embarked_Q']=0
df.loc[df['Embarked_S']==True,'Embarked_S']=1
df.loc[df['Embarked_S']==False,'Embarked_S']=0
df.loc[df['Embarked_C']==True,'Embarked_C']=1
df.loc[df['Embarked_C']==False,'Embarked_C']=0
```



````python
df
````

````python
#verificando se as outras colunas que não foram trabalhadas tem valor NAN
df[['Survived','Pclass','Age','Parch','Fare']].isna().value_counts()
````

````python
#Ajustando os NAN da idade como a mediana de acordo com o sexo e a Pclass, a retirada dos NAN tambem poderiam ter sido realizadas de diversas outras formas
#Male
df.loc[df.Age.isna() & (df.Pclass == 1) & (df.Sex_le == 'male'), 'Age'] = df[df.Pclass == 1 & (df.Sex_le == 'male')].Age.median()
df.loc[df.Age.isna() & (df.Pclass == 2) & (df.Sex_le== 'male'), 'Age'] = df[df.Pclass == 2 & (df.Sex_le == 'male')].Age.median()
df.loc[df.Age.isna() & (df.Pclass == 3) & (df.Sex_le == 'male'), 'Age'] = df[df.Pclass == 3 & (df.Sex_le == 'male')].Age.median()

#Female
df.loc[df.Age.isna() & (df.Pclass == 1) & (df.Sex_le == 'female'), 'Age'] = df[df.Pclass == 1 & (df.Sex_le == 'female')].Age.median()
df.loc[df.Age.isna() & (df.Pclass == 2) & (df.Sex_le== 'female'), 'Age'] = df[df.Pclass == 2 & (df.Sex_le == 'female')].Age.median()
df.loc[df.Age.isna() & (df.Pclass == 3) & (df.Sex_le == 'female'), 'Age'] = df[df.Pclass == 3 & (df.Sex_le == 'female')].Age.median()

````
